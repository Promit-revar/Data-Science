```python
def sp(i):
    if(type(i)==str):
        
        if(i.lower()=='undisclosed' or i.lower()=='unknown' ):
            return 0.0
        return i.replace(",","")
    else:
        return i        
s=data
s.loc[:,'Amount in USD']=s['Amount in USD'].apply(sp)
s.loc[:,'Amount in USD']=pd.to_numeric(s['Amount in USD'], errors ='coerce')
mean=round(np.mean(s['Amount in USD']),2)
print(mean)
s['Amount in USD'].fillna(mean,inplace=True)
s['Amount in USD']
```

```python
def fill(i):
    
    if i=='Delhi' or i=='Delhi ' or i=='New Delhi ':
       
        i='New Delhi'
        
    if i=='banglore':
        i=i.replace("banglore","Banglore")
    return i
#print(g.index("Delhi"))

data['City  Location']=data['City  Location'].apply(fill)
d=Counter(list(data['City  Location']))
#data
d
```

```python
l=data[data['City  Location']=='New Delhi'].iloc[:,8]
l.sum()
py=dict()
for i in d.keys():
    l=data[data['City  Location']==i].iloc[:,8]
    
    py[i]=l.sum()
s=sorted(py.items(),key=lambda x:x[1],reverse=True)
s
```

```python
l=[]
labels=[]
for i in range(10):
    l.append(s[i][1])
    labels.append(s[i][0])
plt.figure(figsize=(10,8))
plt.pie(l,labels=labels,wedgeprops={'edgecolor':"black"},autopct='%1.2f%%',shadow=True)
plt.show()
```

# Investment type

```python
data.columns=['Sr No', 'Date dd/mm/yyyy', 'Startup Name', 'Industry Vertical','SubVertical', 'City  Location', 'Investors Name', 'Investment Type','Amount in USD', 'Remarks']
data.head(10)
```

```python
def clean(i):
    s=i.lower()
    m=0
    if(s.find("private")==0 or s.find("equity")):
        i="Private Equity"
        m=1
    if(s.find("seed")==0):
        i="Seed Funding"
        m=1
    if(s.find("debt")==0):
        i="Debt Funding"
        m=1
    if(s.find("crowd")==0):
        i="Crowd Funding"
        m=1
    if(m==0):
        n=np.random.randint(0,1000)%4
        #print(n)
        l=["Private Equity","Seed Funding","Debt Funding","Crowd Funding"]
        i=l[n]
    
    return i
#print(data['Investment Type'])
data['Investment Type']=data['Investment Type'].astype(str)
print(data['Investment Type'])
data['Investment Type']=data['Investment Type'].apply(clean)
```

```python
b=["Private Equity","Crowd Funding","Seed Funding","Debt Funding"]
l=[]

for i in b:
    x=data[data['Investment Type']==i]
    l.append(x['Amount in USD'].sum())
plt.figure(figsize=(10,8))
plt.pie(l,labels=b,wedgeprops={"edgecolor":"yellow"},autopct='%1.2f%%')
plt.show()
```

# Funding Rounds

```python
data.head()
```

```python
def clean_name(i):
        i=i.lower()
        if(i.find('ola')>=0):
            i="Ola"
        if i.find('oyo')>=0:
            i='Oyo'
        if i.find('flip')>=0:
            i='Flipkart'
        s=i.split(".")
        
        if(len(s)>=3):
            i=s[1]
            
        else:
            i=s[0]
        return i
data['Startup Name']=data['Startup Name'].astype(str)
data['Startup Name']=data['Startup Name'].apply(clean_name)
data['Startup Name'].describe()
```

```python
l=Counter(list(data['Startup Name']))
l=sorted(l.items(),key=lambda x:x[1],reverse=True)
top_five=l[:5]
m=[i[0] for i in top_five]
funds=[]
for i in m:
    funds.append(data[data['Startup Name']==i]['Amount in USD'].sum())
m,funds
```

```python
plt.figure(figsize=(10,8))
plt.pie(funds,labels=m,wedgeprops={"edgecolor":"b"},autopct='%1.2f%%')
plt.show()
f=np.array(funds,dtype=float)
s=f.sum()
k=((f/s)*100).round(2)
for i in range(5):
    print(m[i]+" "+str(k[i]))
```

# Top Industries

```python
data.head()
```

```python
def insclean(i):
    i=i.lower()
    i=i.replace('\\\n',' ')
    #print(i.find('commerce'))
    if(i.find('commerce')>=0 and i.find('e')>=0):
        i='Ecommerce'
    
    return i
#print(data['Industry Vertical'].isnull().sum())
data['Industry Vertical'].fillna(method='ffill',inplace=True)

data['Industry Vertical']=data['Industry Vertical'].astype(str)
data['Industry Vertical']=data['Industry Vertical'].apply(insclean)
l=Counter(list(data['Industry Vertical']))
l=sorted(l.items(),key=lambda x:x[1],reverse=True)
funds=[]
for i in l[:5]:
    funds.append(data[data['Industry Vertical']==i[0]]['Amount in USD'].sum())
f=np.array(funds)
s=f.sum()
pct=((f/s)*100).round(2)
for i in range(5):
    print(l[i][0]+" "+str(pct[i]))
```
