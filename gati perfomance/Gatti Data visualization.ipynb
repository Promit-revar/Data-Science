```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Kiran Rever grade sheet-2020-21 - Sheet1.csv")
df.head()
```

```python
def clean(i):
    if(i=='NT'):
        i=0.0
  
    return i
```

```python
def fill(i):
    if(i==0.0):
        i=np.mean(df['Marks2'])
        #print(i)
   
    return i
```

```python
df.columns=['SNo', "Name", 'School', 'GrSec', 'Gr', 'Sec',
       'Slot timings', 'Day', 'New Category', 'TName', 'Activity1',
       'Marks', 'Activity2', 'Marks1', 'Activity3', 'Marks2', 'Activity4',
       'Marks3', 'Activity5', 'Marks4', 'Activity6', 'Marks5']



#df['Marks2']=df['Marks2'].dropna(inplace=True)


df['Marks2']=df['Marks2'].apply(clean)
print(df['Marks2'].isnull().sum())
df['Marks2'].astype('float')


df['Marks2']=df['Marks2'].astype("float")
np.mean(df['Marks2'])
```

```python
df.head()
```

```python

df['Marks2']=df['Marks2'].apply(fill)
df['Marks2'].fillna(0,inplace=True)
for i in df['Marks2']:
    print(type(i),i)
```

```python
marks2=list(df['Marks2'])
Names=list(df['Name'])
plt.figure(figsize=(15,30))
plt.barh(Names,marks2)
plt.show()
```

```python
from collections import Counter
marks2
a=[]


for i in range(len(marks2)):
    if(marks2[i]>20):
        a.append(">20")
    elif(marks2[i]>15):
        a.append(">15")
    elif(marks2[i]>0 and marks2[i]<=10):
        a.append("<11")
    elif(marks2[i]==0):
        a.append("Not attemped")
    
x=Counter(a)
#
labels=list(x.keys())
cols=['#ff0013','#ffd044','#12ff13','#a023fd']
plt.figure(figsize=(15,12))
plt.pie(list(x.values()),labels=labels,wedgeprops={'edgecolor':'yellow'},explode=[0,0,0],colors=cols,shadow='True',startangle=90,autopct='%1.1f%%')
plt.title("Marks PieChart",color='#ff0000')
plt.show()
```
