## Heatmap
### heatmap can be used for exploratory analysis where we can see how much a feature is correlated to another feature then also we can find the max values or what values does a particular column usually takes and etc

```python
data.head()
```

```python
#getting all numeric columns...
numeric=data.select_dtypes(include=[np.number])
numeric
```

```python
categorical=data.select_dtypes(include=[np.object])
categorical
```

```python
# now we find correlation between numeric data...
numeric_correlation=numeric.corr()
numeric_correlation
```

```python
#ploting heatmap to check correlated features...
sns.heatmap(data=numeric_correlation,annot=True,vmin=0.3,cmap='viridis')
plt.show()
```

### The insights we can gain out of this is that light colored boxes have higher correlation thus we can select them as features if the other value is our target variable and dark color represent high negative correlation...

```python
sns.heatmap(data=numeric)
plt.show()
```
