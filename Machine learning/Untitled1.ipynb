```python
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=datasets.load_boston()
features=data.feature_names
tar=data.target
X=data.data
data=pd.DataFrame(X)
data.columns=features
#data['Price']=tar
data
```

```python
#adding features..
data1=data.copy()

for i in features:
    data[i+'_'+i]=data[i]**2
for i in range(features.size):
    for j in range(i+1,features.size):
        data1[features[i]+'_'+features[j]]=data1[features[i]]*data1[features[j]]
data1= pd.concat([data1, data], axis=1)
data1.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(data1,tar,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
algo=LinearRegression()
algo.fit(X_train,y_train)   
```

```python
pred=algo.predict(X_test)
predt=algo.predict(X_train)
scoret=algo.score(X_train,y_train)
score=algo.score(X_test,y_test)
print(score,scoret)
```
