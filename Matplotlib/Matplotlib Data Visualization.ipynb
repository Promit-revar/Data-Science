<h1>Matplotlib:</h1>
<p>Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays. Matplotlib is written in Python and makes use of NumPy, the numerical mathematics extension of Python. It provides an object-oriented API that helps in embedding plots in applications using Python GUI toolkits such as PyQt, WxPythonotTkinter. It can be used in Python and IPython shells, Jupyter notebook and web application servers also.

Matplotlib has a procedural interface named the Pylab, which is designed to resemble MATLAB, a proprietary programming language developed by MathWorks. Matplotlib along with NumPy can be considered as the open source equivalent of MATLAB.

Matplotlib was originally written by John D. Hunter in 2003.</p>

<h1>format string</h1>
<p>A format string consists of a part for color, marker and line:

fmt = '[marker][line][color]'
Each of them is optional. If not provided, the value from the style cycle is used. Exception: If line is given, but no marker, the data will be a line without markers.

Other combinations such as [color][marker][line] are also supported, but note that their parsing may be ambiguous.</p>

```python
plt.bar(x,y,label="pydev")
plt.bar(x,z,label="jsdev")
plt.legend()
plt.show()
```

```python
x_index=np.arange(len(x))
width=0.2
plt.bar(x_index+width,y,width=width,label="pydev")
plt.bar(x_index-width,z,width=width,label="jsdev")
plt.xticks(ticks=x_index,labels=x)     #ploting for x_index but labeling x values...
plt.legend()
plt.show()
```

<h1>Pie Chart</h1>

```python
Marks=[24,28,27,26.5,25]
labels=['TOC','Java','OS','ALA','CAO']
plt.pie(Marks,labels=labels)
plt.show()
```

```python
plt.pie(Marks,labels=labels,wedgeprops={'edgecolor':'yellow'})
plt.show()
```

```python
plt.pie(Marks,labels=labels,wedgeprops={'edgecolor':'yellow'},colors=['#12ff13','#ffd044','#bcff3a','#a023fd','#123456'])
plt.show()
```

```python
cols=['#12ff13','#ffd044','#bcff3a','#a023fd','#ff3412']
plt.pie(Marks,labels=labels,wedgeprops={'edgecolor':'yellow'},colors=cols,explode=[0,0.1,0,0,0],shadow='True',startangle=90,autopct='%1.1f%%')
plt.title("Marks PieChart",color='#ff0000')
plt.show()
```

```python
data=pd.read_csv('amazon_jobs_dataset.csv')
data.head()
```

```python
from collections import Counter
s=list()
for i in data['Posting_date']:
    s.append(int(i[len(i)-4:len(i)]))
c=Counter(s)
list(c.keys())
```

```python
x=list(c.keys())
y=list(c.values())
plt.plot(x,y)
plt.xlabel('Years')
plt.ylabel('Job Postings')
plt.show()
```
