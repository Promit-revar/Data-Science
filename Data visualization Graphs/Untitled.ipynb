## Barplot

```python
data2=pd.read_csv("data2.csv")
x=list(data2['male'])
y=list(data2['female'])
z=list(data2['postal'])
print(x[0],y[0],z[0])
for i in range(len(data2['state'])):
    plt.bar(["male","Female","Postal"],[x[i],y[i],z[i]],color=["red","blue","orange"])
    plt.xlabel("Voters")
    plt.ylabel("No. of voters")
    plt.title(data2['state'][i])
    plt.show()
```

## Pie chart

```python
cols=['#52f6f3','#bcff3a','#ff3412']

for i in range(len(data2['state'])):
    plt.pie([x[i],y[i],z[i]],labels=["Male Voters","Female Voters","Postal Voters"],colors=cols,wedgeprops={'edgecolor':'yellow'},startangle=90,autopct='%1.1f%%')
    plt.title(data2['state'][i],color='#ff0000')
    plt.show()
```

## Line plot

```python
plt.plot(x,y,'ok--',label="Y-data") 
#plt.plot(x,z,'.-r',label="Z-data") 
plt.legend()
plt.tight_layout()
plt.xlabel("Age")
plt.ylabel("Salary (USD)")
plt.grid(True)
plt.show()
```
