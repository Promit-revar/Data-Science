# There are some preexisting datasets in seaborn...

```python
sns.get_dataset_names()
```

```python
#loading iris data set...
data=sns.load_dataset('tips') #This function returns a dataFrame
data
```

## 1) Statistical visualization...

```python
fig, scatter = plt.subplots(figsize = (11,7))
scatter=sns.scatterplot(x='total_bill',y='tip',data=data)
plt.show()
```

### Here we can infer that there is some relation between total bill and tip higher the total bill greater is the tip...

```python

sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',hue='time',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',hue='time',col='smoker')
plt.show()
```

# 2) Categorical visualization
## Catplot()
## Scatterplots:
### 1) stripplot()

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(x='sex',y='total_bill',data=data,jitter=0.5)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=data,jitter=0.05,hue='smoker',kind='strip')
plt.show()
```

### 2) Swarmplot()

```python
sns.catplot(x='sex',y='total_bill',data=data,hue='smoker',kind='swarm')
plt.plot()
```

## Distributionplots:
### 1) boxplot()

```python
data= sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(x='sex',y='total_bill',data=data,kind='box')
plt.show()
```

### We can gain insight from this plot as more values for male are above average i.e male group comes than chances are that they will have bill greater than average (18) Whereas fro female group it will ponder around average bill for female (15)

```python
sns.catplot(x='sex',y='total_bill',data=data,kind='box',hue='smoker')
plt.show()
```

### Smoker males and females pay more...

### 2) Boxen plot

```python
sns.catplot(x='sex',y='total_bill',data=data,kind='boxen',hue='smoker')
plt.show()
```

### More quartiles are shown...

### 3) Violin plots

```python
data=sns.load_dataset('iris')
data.head()
```

```python
sns.catplot(x='species',y='sepal_width',data=data,kind='violin')
plt.show()
```

### Here we can infer that for setosa mostly sepal width is around average i.e 3.3 or 3.4 
### For Versicolor it is nearly above average (2.7) and is at 2.8 or 2.9 has max density
### finally for virginica it is slightly below average (3.0) it stands at 2.9 to 2.95

# 2) Categorical estimate plots
## Catplot()
### 1) barplot()

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
sns.catplot(x='who',y='fare',data=data,kind='bar')
plt.show()
```

```python
sns.catplot(x='who',y='fare',data=data,kind='bar',hue='class',palette='Blues')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=data,hue='class',estimator = np.median,ci = 0)
plt.show()
```

```python
sns.catplot(x = 'class', y = 'fare', data = data,order = ["Third", "Second", "First"],kind='bar',estimator=np.mean)
plt.plot()
```

### 2) Pointplot()

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.pointplot(x="time", y="total_bill", data=data)
plt.show()
```

```python
sns.catplot(x="time", y="total_bill",kind='point',hue="smoker",data=data, dodge=True,estimator=np.mean)
plt.plot()
```

```python
sns.catplot(x="time", y="total_bill",kind='point',hue="smoker",data=data, dodge=True,estimator=np.median)
plt.plot()
```

### 3) countplot()

```python
data=sns.load_dataset('mpg')
data.head()
```

```python
sns.catplot(x='origin',data=data,kind='count')
plt.plot()
```

```python
sns.countplot(x='origin',hue='model_year',data=data)
plt.show()
```
