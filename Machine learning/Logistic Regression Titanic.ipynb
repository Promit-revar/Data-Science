```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import math
```

```python
datacheck=sns.load_dataset('titanic')
datacheck
```

```python
X_train=pd.read_csv("train.csv")
X_test=pd.read_csv("test.csv")


X_train.drop(['Name'],axis=1,inplace=True)
X_train['Sex']=X_train['Sex'].rank(method='dense', ascending=False).astype(int)
avgAge=round(np.mean(X_train['Age']),0)
X_train['Age']=X_train['Age'].fillna(avgAge)
#X_train['Age']=X_train['Age'].dropna()
X_train['Ticket']=X_train['Ticket'].rank(method='dense', ascending=False).astype(int)

X_train.dropna(inplace=True)
X_train['Embarked']=X_train['Embarked'].rank(method='dense', ascending=False).astype(int)
X_train['Cabin']=X_train['Cabin'].rank(method='dense', ascending=False).astype(int)
Y_train=X_train['Survived']
X_train.drop('Survived',axis=1,inplace=True)
X_train
```

```python
#X_test.drop('Cabin',axis=1,inplace=True)
X_test['Sex']=X_test['Sex'].rank(method='dense', ascending=False).astype(int)
avgAge=round(np.mean(X_test['Age']),0)
X_test['Age']=X_test['Age'].fillna(avgAge)
#X_train['Age']=X_train['Age'].dropna()
X_test['Ticket']=X_test['Ticket'].rank(method='dense', ascending=False).astype(int)
#X_test['Embarked']=X_test['Embarked'].rank(method='dense', ascending=False).astype(int)
X_test.dropna(inplace=True)
X_test['Embarked']=X_test['Embarked'].rank(method='dense', ascending=False).astype(int)
X_test.drop("Name",axis=1,inplace=True)
X_test['Cabin']=X_test['Cabin'].rank(method='dense', ascending=False).astype(int)
X_test
```

```python
clf=LogisticRegression()
clf
```

```python
clf.fit(X_train,Y_train)
```
