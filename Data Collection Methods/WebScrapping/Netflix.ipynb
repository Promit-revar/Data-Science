```python
import requests as req
```

```python
data=req.request("GET","https://www.netflix.com/browse/genre/58806?bc=34399")
data
html=data.text
```

```python
from bs4 import BeautifulSoup 
```

```python
x=BeautifulSoup(html,'html.parser')
print(x.prettify())
```

```python
l=x.find('script').string
```

```python
import json
```

```python
movies=json.loads(l)
movies
```

**Converting to CSV**

```python
import pandas as pd
d={}
for i in movies['itemListElement']:
    d.update({i['position']:i['item']})
    #print(d)
data=pd.DataFrame(d).T
#movies['itemListElement'][0]['item']['name']
#data
data.columns=['type','Name','URL']
data.to_csv("./Movie_on_netflix.csv")
```
