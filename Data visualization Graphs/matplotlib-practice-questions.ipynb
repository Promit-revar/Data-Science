```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('amazon_jobs_dataset.csv')
data.head()
```

```python
data.describe()
```

```python
#data cleaning...
data[data['PREFERRED QUALIFICATIONS'].isnull()==True]
data['PREFERRED QUALIFICATIONS'].fillna("Any",inplace=True)
data[data['BASIC QUALIFICATIONS'].isnull()==True]
data['BASIC QUALIFICATIONS'].fillna("Any",inplace=True)
data[data['DESCRIPTION'].isnull()]
data['DESCRIPTION'].fillna("Not described",inplace=True)
data.isnull().sum()
data.shape[0]
```

```python
#Month VS Job opennings...
from collections import Counter
l=list(data["Posting_date"])

months=[]
for i in l:
    k=i.split()
    months.append(k[0])
#print(months)
s=Counter(months)
plt.barh(list(s.keys()),list(s.values()))
plt.xlabel("Number of Job Openings")
plt.ylabel("Months")
plt.show()
```

```python
#data2=data[data['location'][0:][:2]=='IN']
cities=[]
for i in data['location']:
    if(i[:2]=="IN"):
        k=i.split(',')
        cities.append(k[2][1:len(k[2])-1])
cts=Counter(cities)
lbs=cts.keys()
cols=['#12ff13','#ffd044','#bcff3a','#a023fd','#ff3412']
plt.pie(cts.values(),colors=cols[:len(cols)-1],wedgeprops={'edgecolor':'yellow'},autopct='%1.2f%%',labels=lbs,shadow="True",startangle=90)
plt.show()
```

```python
x=list(data['Title'])
data2=list(data["Posting_date"])
desired=[]
year=[]
for i in range(len(x)):
    if "java" in x[i].lower():
        desired.append(x[i])
        s=data2[i].split(',')
        year.append(s[1][1:])

year.sort()
plt.scatter(year,desired)


plt.xlabel("Year")
plt.ylabel("Java Job openings")
plt.show()
```
